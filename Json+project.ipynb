### 1. Find the top 10 countries with the most projects

To locate the top 10 countries with the most projects, we need to run a ".value_counts" method on the 'countryname' column.

### 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
df.head(15)
```

Looking at the Dataframe, it appears that the 'theme_namecode' column contains nested data. We will need to normalize the column in order to find the top 10 major project theme codes.

```python
# Load json file as string to normalize 'mjtheme_namecode' column :
df2 = json.load(open(r'C:\Users\eshai\Desktop\Data Science Material\Python Scripts\data_wrangling_json\data\world_bank_projects.json'))
```

```python
#normalize 'mjtheme_namecode' column and insect :
df2 = json_normalize(df2, 'mjtheme_namecode')

df2.head(10)
```

Now that the 'mjtheme_namecode' column has been normalized, we can count the values within the column to determine the Top 10 Major project theme codes

```python
# count the values in 'code' column :
Top_10_theme = df2.code.value_counts().head(10)

# print counted values in column :
Top_10_theme
```

 The list above show cases the top 10 Theme codes within the dataframe

###  3. Create a dataframe with the missing names filled in.

 First we need to sort the dataframe in order to get a better understanding of what values are missing within the 'name' column.

```python
# sort values in 'df2' by 'code' and 'name':
df2 = df2.sort_values(['code', 'name'])

# print sorted data:
df2
```

 We will need to replace the empty cells with Nan values in order to fill those missing values within the 'name' column

```python
# fill blank entries in 'name' column with nan :
df2.name[df2['name'] == ''] = np.nan

# Print 'df2' to view changes entries:
df2.head(15)
```

Now that the empty cell have been replaced by Nan values, we can use the "fillna()" method to backfill name values based on the entry below it to fill in the appropriate project names.

```python
# Backfill and replace vaules:
df2 = df2.name.fillna(method='bfill')

# view df2:
df2.head(15)
```

Finally, we can use the same value count method as we did earlier to count the project code names and show the Top 10 values.

```python
# count values of complete 'name' column :
top_10_themes = df2.value_counts().head(10)

#view Top 10 values :
top_10_themes
```

Lastly, the list above shows that the Top 10 Project names in the dataframe. Notice the value count for each key name is consistent with the value count for the 'mjtheme_namecode' column proving that our fill method was done correctly.
